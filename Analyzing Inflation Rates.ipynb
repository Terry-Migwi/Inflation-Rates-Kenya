### 2. Understanding the Context

Inflation is the rate at which the value of a currency is falling and consequently the general level of prices for goods and services in a country. Inflation measures the annual percentage rise in the cost of living. A rise in inflation rates means the price of goods and services are rising at a faster rate hence affecting the economy of a country negatively. 

There are various factors that affect inflation rates of a country. This study will consider certain income levels and the consumer price indices to Identifying which of these factors affect rates of inflation and consumer price indices hence we will obtain insights on how any changes made on this factors will impact the economy.

```python
# confirming high correlation observed from the report

def correlation_heatmap(full_dataset):
    correlations = full_dataset.corr()

    fig, ax = plt.subplots(figsize=(10,10))
    sns.heatmap(correlations, vmax=1.0, center=0, fmt='.2f',
                square=True, linewidths=.5, annot=True, cbar_kws={"shrink": .70})
    plt.show();
    
correlation_heatmap(full_dataset)
```

Variables with high correlation(correlation > 0.60) include:

* lower income is highly correlated to nairobi combined, rest of urban areas and kenya cpi
* middle income is highly correlated to upper income, nairobi combined, rest of urban areas, kenya cpi and overall cpi
* upper income is highly correlated to middle income, rest of urban areas and kenya cpi

* rate of inflation is not highly correlated to any variable. It has a moderate negative correlation to lower income, and a low negative correlation to nairobi combined and kenya cpi.

**Conclusions**
